```python
from bs4 import BeautifulSoup
import requests
url = 'https://www.spoilertime.ru/index.php/films/1582-tom-i-dzherri'
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')

print(soup.prettify())
```

```python
# Заголовок
soup.section.ul.find_all('li', {"class": "active"}, itemprop="itemListElement")[-1].get_text().strip()
```

```python
# Тело_1
soup.find('div', {"style": "overflow: hidden;padding-bottom: 20px;"}).get_text().strip()
```

```python
# Тело_2
soup.find_all('div', {"class": "leading-1"})[0].get_text().strip()
```
